```python
# Grouping by trim and calculating average selling price for each trim
average_prices = filtered_data.groupby('trim')['sellingprice'].mean()

# Sorting the average prices in descending order and selecting the top 5
top_5_trims = average_prices.sort_values(ascending=False).head(5)

# Plotting the top 5 highest average selling prices based on trim
top_5_trims.plot(kind='bar', color='skyblue')
plt.title('Top 5 Highest Average Selling Prices by Trim')
plt.xlabel('Trim')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45)
plt.show()
```

```python
trim_counts = filtered_data['trim'].value_counts()
popular_trims = trim_counts[trim_counts > 100].index
df_filtered = filtered_data[filtered_data['trim'].isin(popular_trims)]

# Grouping by trim and calculating average selling price and count for each trim
summary_df = df_filtered.groupby('trim').agg({'sellingprice': 'mean', 'trim': 'count'})
summary_df.columns = ['average_selling_price', 'count']

# Sorting by average selling price in descending order and selecting the top 5
top_50_summary = summary_df.sort_values(by='average_selling_price', ascending=False).head(20)

# Plotting the top 5 highest average selling prices and counts based on trim
fig, ax1 = plt.subplots(figsize=(15, 6))

color = 'navy'
ax1.set_ylabel('Average Selling Price', color=color)
ax1.bar(top_50_summary.index, top_50_summary['average_selling_price'], color=color)
ax1.set_xticklabels(top_50_summary.index, rotation=45, ha='right', rotation_mode='anchor')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'orange'
ax2.set_ylabel('Count', color=color)
ax2.plot(top_50_summary.index, top_50_summary['count'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Top 10 Highest Average Selling Prices and Counts by Common Trim')
plt.show()
```

```python
# Calculate the count of sedan and SUV
sedan_count = df_filtered[df_filtered['body'] == 'Sedan'].shape[0]
suv_count = df_filtered[df_filtered['body'] == 'SUV'].shape[0]
truck_count = df_filtered[df_filtered['body'] == 'Truck'].shape[0]

# Calculate the count of other body types
other_count = df_filtered.shape[0] - sedan_count - suv_count - truck_count

# Data for the pie chart
labels = ['Sedan', 'SUV', 'Truck', 'Other']
sizes = [sedan_count, suv_count, truck_count, other_count]
colors = ['lightblue', 'lightgreen', 'yellow','lightcoral']
explode = (0.1, 0.1, 0.1, 0.1)  # explode the 1st slice (Sedan)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Sedan, SUV, Truck, and Other Body Types')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter DataFrame for only SUV and Sedan
df_suv_sedan = df_filtered[df_filtered['body'].isin(['SUV', 'Sedan'])]

# Calculate the count of SUV and Sedan
suv_count = df_suv_sedan[df_suv_sedan['body'] == 'SUV'].shape[0]
sedan_count = df_suv_sedan[df_suv_sedan['body'] == 'Sedan'].shape[0]

# Group by 'make' and calculate the percentage
make_counts_suv = df_suv_sedan[df_suv_sedan['body'] == 'SUV']['make'].value_counts(normalize=True)
make_counts_sedan = df_suv_sedan[df_suv_sedan['body'] == 'Sedan']['make'].value_counts(normalize=True)

# Filter makes contributing more than 5% for clarity
make_counts_suv = make_counts_suv[make_counts_suv >= 0.05]
make_counts_sedan = make_counts_sedan[make_counts_sedan >= 0.05]

# Data for the pie charts
labels_suv = make_counts_suv.index
sizes_suv = make_counts_suv.values * suv_count
labels_sedan = make_counts_sedan.index
sizes_sedan = make_counts_sedan.values * sedan_count

# Plotting the pie chart for SUV
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(sizes_suv, labels=labels_suv, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Make within SUV')

# Plotting the pie chart for Sedan
plt.subplot(1, 2, 2)
plt.pie(sizes_sedan, labels=labels_sedan, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Make within Sedan')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter DataFrame for only Truck
df_truck = df_filtered[df_filtered['body'].isin(['Truck'])]

# Calculate the count of Truck
truck_count = df_truck[df_truck['body'] == 'Truck'].shape[0]

# Group by 'make' and calculate the percentage
make_truck = df_truck[df_truck['body'] == 'Truck']['make'].value_counts(normalize=True)

# Filter makes contributing more than 5% for clarity
make_truck = make_truck[make_truck >= 0.05]

# Data for the pie chart
labels_truck = make_truck.index
sizes_truck = make_truck.values * truck_count

# Data for the 'Other' category
other_count = truck_count - int(sizes_truck.sum())
labels_truck = labels_truck.tolist() + ['Other']
sizes_truck = sizes_truck.tolist() + [other_count]

# Plotting the pie chart for Truck
plt.figure(figsize=(8, 8))
plt.pie(sizes_truck, labels=labels_truck, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Make within Truck')
plt.tight_layout()
plt.show()
```

```python
# Grouping by state and calculating average selling price, count, and average MMR for each state
state_summary = df.groupby('state').agg({'sellingprice': 'mean', 'state': 'count', 'mmr': 'mean'})
state_summary.columns = ['average_selling_price', 'count', 'average_mmr']

# Calculate the difference between average MMR and average selling price
state_summary['mmr_difference'] = state_summary['average_selling_price'] - state_summary['average_mmr']

# Filtering states with low competition (count < threshold) and high return (selling price > MMR)
threshold_count = 100000  # Set your threshold for low competition
low_competition_high_return = state_summary[(state_summary['count'] < threshold_count) & (state_summary['average_selling_price'] > state_summary['average_mmr'])]

print(low_competition_high_return)  # Print the filtered dataframe to check its content

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(state_summary['count'], state_summary['average_selling_price'], s=state_summary['average_mmr']/500, c='blue', alpha=0.5)
if not low_competition_high_return.empty:
    plt.scatter(low_competition_high_return['count'], low_competition_high_return['average_selling_price'], s=low_competition_high_return['average_mmr']/500, c='red', label='Low Competition, High Return')

for state, count, price in zip(state_summary.index, state_summary['count'], state_summary['average_selling_price']):
    plt.text(count, price + 200, state, fontsize=9, ha='center', va='bottom')

plt.xlabel('Count of Listings')
plt.ylabel('Average Selling Price')
plt.title('Average Selling Price vs. Count of Listings by State')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Selecting specific columns
new_df = df[['make', 'model','state', 'sellingprice', 'mmr']]

# Filter the DataFrame
new_data_df = new_df[new_df['state'].isin(['on', 'co', 'wa'])]
new_data_df = new_df[new_df['make'].isin(['Ford', 'Jeep', 'Infiniti', 'BMW', 'Toyota', 'Dodge', 'Nissan', 'Hyundai' ,'Kia'])]

# Combine make and model columns
new_data_df['make_model'] = new_data_df.apply(lambda row: f"{row['make']} {row['model']}", axis=1)

# Drop the original make and model columns if needed
new_data_df.drop(columns=['make', 'model'], inplace=True)
```

```python
# Calculate average MMR and selling price
avg_mmr = new_data_df['mmr'].mean()
avg_selling_price = new_data_df['sellingprice'].mean()

# Calculate difference between MMR and selling price
new_data_df['diff'] = new_data_df['mmr'] - new_data_df['sellingprice']

print(new_data_df)
```

```python
# Drop the 'state' column
new_data_df.drop(columns=['state'], inplace=True)

# Group by 'make_model', calculate average MMR and selling price, and count occurrences
avg_data = new_data_df.groupby('make_model').agg({'mmr': 'mean', 'sellingprice': 'mean', 'diff': 'count'}).reset_index()

# Rename the count column to 'count'
avg_data.rename(columns={'diff': 'count'}, inplace=True)

print(avg_data)
```

```python
import matplotlib.pyplot as plt

# Sort avg_data by avg_diff in descending order
top_10_avg_diff = avg_data.sort_values(by='avg_diff', ascending=False).head(10)

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for average difference
ax1.bar(top_10_avg_diff['make_model'], top_10_avg_diff['avg_diff'], color='skyblue')
ax1.set_xticklabels(top_10_avg_diff.make_model, rotation=45, ha='right', rotation_mode='anchor')
ax1.set_ylabel('Average Difference (MMR - Selling Price)', color='blue')
ax1.tick_params('y', colors='blue')
ax1.set_title('Top 10 Highest Average Difference (MMR - Selling Price)')

# Create a twin Axes sharing the xaxis
ax2 = ax1.twinx()

# Line plot for count
ax2.plot(top_10_avg_diff['make_model'], top_10_avg_diff['count'], color='orange', marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Count', color='orange')
ax2.tick_params('y', colors='orange')

# Add count labels to the line plot
for i, value in enumerate(top_10_avg_diff['count']):
    ax2.text(i, value, str(value), ha='center', va='bottom')

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
